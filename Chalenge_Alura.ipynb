# **Importar Datos**

```python
import pandas as pd
import numpy as np

url_1 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv'
url_2 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv'
url_3 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv'
url_4 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv'

pd.read_csv(url_1)
pd.read_csv(url_2)
pd.read_csv(url_3)
pd.read_csv(url_4)
```

```python
df_1 = pd.read_csv(url_1)
df_2 = pd.read_csv(url_2)
df_3 = pd.read_csv(url_3)
df_4 = pd.read_csv(url_4)
df_4.dtypes
```

## **Analicis de información**

### **Calcular ventas totales de cada tienda**

```python
total_1 = df_1['Precio'].sum()
total_2 = df_2['Precio'].sum()
total_3 = df_3['Precio'].sum()
total_4 = df_4['Precio'].sum()

print(f'Total ventas tienda 1: {total_1} \n Total ventas tienda 2: {total_2} \n Total ventas tienda 3: {total_3} \n Total ventas tienda 4: {total_4}')
```

```python
ventas_totales = pd.Series([total_1, total_2, total_3, total_4], index=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])
ventas_totales
```

### **ventas por categoria**

```python
V_categoria_1 = df_1.groupby('Categoría del Producto')['Precio'].count()
V_categoria_2 = df_2.groupby('Categoría del Producto')['Precio'].count()
V_categoria_3 = df_3.groupby('Categoría del Producto')['Precio'].count()
V_categoria_4 = df_4.groupby('Categoría del Producto')['Precio'].count()

datos_V_Categoria = {'Tienda 1': V_categoria_1, 'Tienda 2': V_categoria_2, 'Tienda 3': V_categoria_3, 'Tienda 4': V_categoria_4}

datos_V_Categoria
```

```python
dfs = [df_1, df_2, df_3, df_4]
nombres_tiendas = [f'Tienda {i+1}' for i in range(len(dfs))]

datos_V_Categoria = pd.concat([df.groupby('Categoría del Producto')['Precio'].count() for df in dfs], axis=1)
datos_V_Categoria.columns = nombres_tiendas

datos_V_Categoria
```

### **Satifacción del cliente**

```python
cal_1 = df_1['Calificación'].mean()
cal_2 = df_2['Calificación'].mean()
cal_3 = df_3['Calificación'].mean()
cal_4 = df_4['Calificación'].mean()

print(f'Calificación promedio tienda 1: {cal_1} \n Calificación promedio tienda 2: {cal_2} \n Calificación promedio tienda 3: {cal_3} \n Calificación promedio tienda 4: {cal_4}')
```

```python
promedio_calificacion = pd.Series([cal_1, cal_2, cal_3, cal_4], index=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])
promedio_calificacion
```

### **Productos más Vendidos**

```python
Vproduct_1 = df_1.groupby('Producto')['Precio'].count().pipe(lambda x: x[x == x.max()])
Vproduct_2 = df_2.groupby('Producto')['Precio'].count().pipe(lambda x: x[x == x.max()])
Vproduct_3 = df_3.groupby('Producto')['Precio'].count().pipe(lambda x: x[x == x.max()])
Vproduct_4 = df_4.groupby('Producto')['Precio'].count().pipe(lambda x: x[x == x.max()])

datos = {'Tienda 1': Vproduct_1, 'Tienda 2': Vproduct_2, 'Tienda 3': Vproduct_3, 'Tienda 4': Vproduct_4}

datos


```

```python
Miproducto_1 = df_1.groupby('Producto')['Precio'].count().pipe(lambda x: x[x == x.min()])
Miproducto_2 = df_2.groupby('Producto')['Precio'].count().pipe(lambda x: x[x == x.min()])
Miproducto_3 = df_3.groupby('Producto')['Precio'].count().pipe(lambda x: x[x == x.min()])
Miproducto_4 = df_4.groupby('Producto')['Precio'].count().pipe(lambda x: x[x == x.min()])

datos = {'Tienda 1': Miproducto_1, 'Tienda 2': Miproducto_2, 'Tienda 3': Miproducto_3, 'Tienda 4': Miproducto_4}

datos
#
```

```python
# Función para obtener productos con min y max ventas
def productos_min_max(df):
    conteo = df.groupby('Producto')['Precio'].count()
    return pd.Series({
        'Menor venta': ', '.join(conteo[conteo == conteo.min()].index),
        'Mayor venta': ', '.join(conteo[conteo == conteo.max()].index)
    })

# Aplicar a todas las tiendas
productos_min_max_df = pd.DataFrame(
    [productos_min_max(df) for df in dfs],
    index=nombres_tiendas
)

productos_min_max_df
```

### **Valor de envio promedio por Tienda**

```python
Prom_envio_1 = df_1['Costo de envío'].mean()
Prom_envio_2 = df_2['Costo de envío'].mean()
Prom_envio_3 = df_3['Costo de envío'].mean()
Prom_envio_4 = df_4['Costo de envío'].mean()

print(f'Valor de envio promedio tienda 1: {Prom_envio_1} \n Valor de envio promedio tienda 2: {Prom_envio_2} \n Valor de envio promedio tienda 3: {Prom_envio_3} \n Valor de envio promedio tienda 4: {Prom_envio_4}')
```
